```python
import pandas as pd
import numpy as np
import plotnine as p9
import view_cube as vc
import turn_cube as tc
import random
```

```python
cube = np.zeros((6, 3, 3), dtype=int)

for x in range(6):
    for y in range(3):
        for z in range(3):
            cube[x,y,z]=x

vc.draw(cube)
```
